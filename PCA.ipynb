```python
#reduce 4 features to 2 or 3 by using pca
import pandas as pd
df=pd.read_csv('/content/Iris.csv')
df
```

```python
x=df.iloc[:,1:5].values
x
```

```python
y=df.iloc[:,-1].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)  # here we learn/understand(fit) x train and transform using that values, xtest values only transform
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=2)  # mention how many components should take
x_train=pca.fit_transform(x_train)
x_test=pca.transform(x_test)
```

```python
x_train  # 4 principle components
```

```python
pca.explained_variance_ratio_  # pc1 contains more information
```

```python
from sklearn.ensemble import RandomForestClassifier
clf=RandomForestClassifier()
clf.fit(x_train,y_train)
y_pred=clf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred))
```
